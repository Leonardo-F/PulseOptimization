### 评分器对比

这里主要展示，经过我们优化后的可多进程运行的评分器与原评分器的运行性能对比，主要比较出值是否一致，运行时间是否有显著差异。

可以看到使用原始评分器，耗时 324s，使用多进程评分器，耗时 31s，且评分器输出一致，运行速度提升了 10 倍。

```python
# 导入必备的库
import time 
import numpy as np

# 可多进程运行评分器
from two_transmon_grader import DispersiveCNOTPulseGrader

# 原始评分器
from two_transmon_grader_origin import DispersiveCNOTPulseGrader_Origin

# 加载脉冲
pulses = np.load('results/pulses_closed.npy') 
```

```python
# 原始评分器
grader_origin = DispersiveCNOTPulseGrader_Origin(
    nq_levels=3,
    n_steps=300,
    dt=5e-10,
    n_shots=10
)

time_origin = time.time()
# 评分
results = grader_origin.grade_submission(pulses, seed=42, verbose=True)
time_origin = time.time() - time_origin
print(f"原始评分器耗时：{time_origin}s")
```

```python
# 多进程评分器
grader = DispersiveCNOTPulseGrader(
    nq_levels=3,
    n_steps=300,
    dt=5e-10,
    n_shots=10,
    computing_method='parallel'
)

time_1 = time.time()
# 评分
results = grader.grade_submission(pulses, seed=42,verbose=True)
time_1 = time.time() - time_1
print(f"多进程评分器耗时：{time_1}s")
```
